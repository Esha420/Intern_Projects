```python
# Creating Array in numpy
import numpy as np
import math
arr1 = np.array([1,2,3,4,5,6,7])
arr2 = np.array([8,9,10,11,12,13,14])
num1 = 10
num2 = 12
print(arr1)
```

```python
# Operations in Numpy Array
## Binary Operations : Acts on bits and performsbit by bit operations
print("Input number 1 is:", num1)
print("Input number 2 is:", num2)
out_num1 = np.bitwise_and(num1, num2)
out_num2 = np.bitwise_or(num1, num2)
out_num3 = np.bitwise_xor(num1, num2)
out_num4 = np.invert(num1)
print("Bitwise AND OPeration is:", out_num1)
print("Bitwise OR OPeration is:", out_num2)
print("Bitwise XOR OPeration is:", out_num3)
print("Bitwise inversion OPeration is:", out_num4)
```

```python
# Numpy Mathematical Functions
## Trigonometric Functions
sine_values = np.sin(arr1) # gives sin value
print(sine_values)
cosine_values = np.cos(arr1) # gives cos value
print(cosine_values)
inv_sin = np.arcsin(sine_values) # gives sin inverse
print(inv_sin)
inv_cos = np.arccos(cosine_values) # gives cos inverse
print(inv_cos)
log_fun = np.log(arr1) #Log Function
print(log_fun)
cube_root = np.cbrt(num1) #Cube Root
print(cube_root)
clipping = np.clip(arr1, a_min = 2, a_max=4)
print(clipping)
```

```python
# Arthematic Operations
arr2 = arr2.reshape(1,7)
print(np.add(arr1,arr2)) # Adding array
print(np.subtract(arr1,arr2)) # Subtract Array
print(np.multiply(arr1,arr2)) # Multiply Array
print(np.divide(arr1, arr2)) # Divide Array
print(np.power(arr1,2)) # Power to 2
print(np.sum(arr1)) # sum of arr1
print(np.inner(arr1,arr2)) # Computes inner product
```

```python
## Matrix
x = np.array([[1, 2], [4, 5]]) 
y = np.array([[7, 8], [9, 10]]) 
print (np.add(x,y)) # matrix add
print (np.subtract(x,y)) # matrix subtract
print (np.divide(x,y)) # matrix divide
print (np.multiply(x,y)) # element-wise multiply
print (np.dot(x,y)) # dot product
```

```python
# Convert to string and fill NaNs with empty string
df['Stars'] = df['Stars'].astype(str).fillna('')

# Now split and explode safely
all_stars = df['Stars'].str.split(', ').explode()

# Proceed with counting
from collections import Counter
star_counts = Counter(all_stars)
top_stars = star_counts.most_common(10)

# To DataFrame for plotting
import pandas as pd
star_df = pd.DataFrame(top_stars, columns=['Star', 'Episodes'])

# Plot
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.barh(star_df['Star'], star_df['Episodes'], color='plum')
plt.title('Top 10 Most Frequent Stars')
plt.xlabel('Number of Episodes')
plt.gca().invert_yaxis()
plt.grid(axis='x')
plt.show()
```

```python
## Average Votes by Star
star_votes = df[['Votes', 'Stars']].copy()
star_votes['Stars'] = star_votes['Stars'].str.split(', ')

# Explode to have one row per star
exploded = star_votes.explode('Stars')

# Group by star
avg_votes = exploded.groupby('Stars')['Votes'].mean().sort_values(ascending=False).head(10)

avg_votes.plot(kind='barh', color='teal', figsize=(10,6), title='Top 10 Stars by Avg Votes')
plt.xlabel('Average Votes')
plt.gca().invert_yaxis()
plt.grid(axis='x')
plt.show()
```

```python
## Time based Trends
# Votes over years
votes_by_year = df.groupby('Year_of_prod')['Votes'].mean()

plt.figure(figsize=(10, 6))
votes_by_year.plot(marker='o', linestyle='-', color='tomato')
plt.title('Average Votes by Year')
plt.xlabel('Year')
plt.ylabel('Average Votes')
plt.grid()
plt.show()
```

```python
## Duration over Time
duration_trend = df.groupby('Year_of_prod')['Duration'].mean()

plt.figure(figsize=(10, 3))
duration_trend.plot(marker='o', linestyle='--', color='seagreen')
plt.title('Average Duration Over Years')
plt.xlabel('Year')
plt.ylabel('Duration (minutes)')
plt.grid()
plt.show()
```

```python
import re
from collections import Counter

# Extract capitalized words (likely names)
def extract_names(text):
    return re.findall(r'\b[A-Z][a-z]{2,}\b', text)

# Apply to all summaries
name_mentions = df['Summary'].dropna().apply(extract_names)

# Flatten the list of lists
all_names = [name for sublist in name_mentions for name in sublist]

# Count frequency
name_counts = Counter(all_names)
top_names = name_counts.most_common(10)

# Display top 10 names
char_df = pd.DataFrame(top_names, columns=['Name', 'Mentions'])
```

```python
# Step 1: Filter the character names (already done above)
main_characters = {'Rachel', 'Monica', 'Chandler', 'Ross', 'Joey', 'Phoebe', 'Emily'}
filtered_df = char_df[char_df['Name'].isin(main_characters)].reset_index(drop=True)

# Step 2: Create the 'Name_Mentions' column as done before
df['Name_Mentions'] = name_mentions

# Explode the 'Name_Mentions' to get one row per character per episode
exploded = df.explode('Name_Mentions')

# Step 3: Calculate the average votes for each character mentioned in an episode
name_vote_avg = exploded.groupby('Name_Mentions')['Votes'].mean().sort_values(ascending=False)

# Step 4: Merge the name mentions with the average votes
final_df = filtered_df.merge(name_vote_avg, left_on='Name', right_index=True)
final_df.columns = ['Name', 'Mentions', 'Avg_Votes']
final_df.sort_values(by='Mentions', ascending=False, inplace=True)
```
